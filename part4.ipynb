### **Exercise 39**

The 39.csv file contains predictions for a certain regression model:

* y_true describes the actual, observed values

* y_pred describes the values predicted by model

Load this file into a DataFrame. Then implement a function called mean_absolute_error() that computes the mean absolute error.

```python
pred = pd.read_csv('39.csv')
pred.head()
```

```python
def mean_absolute_error(y_true, y_pred):
    return abs(y_true - y_pred).sum() / len(y_true)
 
mae = mean_absolute_error(pred['y_true'], pred['y_pred'])
print(f'MAE = {mae:.4f}')
```

### **Exercise 40**

Implement a function called mean_squared_error() that computes the mean squared error

```python
def mean_squared_error(y_true, y_pred):
    return abs((y_true - y_pred)**2).sum() / len(y_true)
 
mse = mean_squared_error(pred['y_true'], pred['y_pred'])
print(f'MSE = {mse:.4f}')
```
